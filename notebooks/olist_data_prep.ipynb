## 1. Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## 2. Load Dataset

```python
orders = pd.read_csv("../olist_orders_dataset.csv")
items = pd.read_csv("../olist_order_items_dataset.csv")
products = pd.read_csv("../olist_products_dataset.csv")
translation = pd.read_csv('../product_category_name_translation.csv')

products = products.merge(translation, on='product_category_name', how='left')
```

```python
products = products.drop(columns='product_category_name')
```

## 3. Dataset Overview

```python
print("Orders Shape:", orders.shape)
print("Order Items Shape:", items.shape)
print("Product Shape:", products.shape)
```

## 4. Convert Datetime Columns

```python
date_cols = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_cols:
    orders[col] = pd.to_datetime(
        orders[col]
    )
```

```python
items["shipping_limit_date"] = pd.to_datetime(items["shipping_limit_date"])
```

## 5. Export Data

```python
orders.to_csv('../data/cleaned_orders.csv', index=False)
products.to_csv('../data/cleaned_products.csv', index=False)
items.to_csv('../data/cleaned_items.csv', index=False)
```
